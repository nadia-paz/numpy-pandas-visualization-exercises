### Matplotlib exercise Big-O

```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

* Title your chart "Big O Notation"
* Label your x axis "Elements"
* Label your y axis "Operations"
* Label your curves or make a legend for the curves
* Use LaTex notation where possible

**Curves to graph:**
* $y = 0n + 1$ and label the curve "O(1)"
* $y = log(n)$and label the curve "O(log n)"
* $y = n$ and label the curve "O(n)"
* $y = n * log(n)$ and label it "O(n log n)"
* $y = n^2$ and label it "O(n^2)"
* $y = 2^n$ and label it "O(2^n)"
* $y = n!$ and label it "O(n!)"
* $y = n^n$ and label it "O(n^n)"

```python
x = list(range(1, 11))
x6 = list(range(1, 8))
x7 = list(range(1, 6))
x8 = [1, 2, 3, 4]

y1 = [(0 * n + 1) for n in x]
y2 = [np.log10(n) for n in x]
y3 = [n for n in x]
y4 = [(n * np.log10(n)) for n in x]
y5 = [n ** 2 for n in x]
y6 = [2 ** n for n in x6]
y7 = [math.factorial(n) for n in x7]
y8 = [n ** n for n in x8]
```

```python
plt.figure(figsize = (10, 8))
plt.title('Big O Notation')
plt.xlabel('Elements', size = 'large')
plt.ylabel('Operations', size = 'large')
plt.grid(c='grey', alpha = 0.4)
plt.ylim(0, 120)

plt.plot(x, y1,color = 'r', ls = ':', label = 'O(1)')
plt.plot(x, y2,color = 'b', ls = ':', label = 'O($log(n)$)')
plt.plot(x, y3,color = 'g', ls = ':', label = 'O(n)')
plt.plot(x, y4,color = 'm', ls = ':', label = 'O($n(log(n)$)')
plt.plot(x, y5,color = 'firebrick', ls = '--', label = 'O($(n^2$)')
plt.plot(x6, y6,color = 'b', ls = '--', label = 'O($2^n$)')
plt.plot(x7, y7,color = 'r', ls = '--', label = 'O($n!$)')
plt.plot(x8, y8,color = 'g', ls = '--', label = 'O($n^n$)')
plt.legend();
```
