**1. Use matplotlib to plot the following equation:**

**$y = x^2 - x + 2$** 

You'll need to write the code that generates the x and y points. 

Add an anotation for the point 0, 0, the origin.*

```python
#plt.figure(figsize = (6, 8))
x = np.linspace(-10, 10, 100)
y = [x**2 - x + 2 for x in x]
plt.plot(x, y)
plt.title('$y = x^2 - x + 2$')
plt.xlim(-11, 11)
plt.ylim(0, 100)
plt.annotate('Origin', xy=(0,0), xytext=(0, 20), arrowprops = {})
plt.savefig('figure1')
plt.show()
```

###### **2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):**

* $y = \sqrt{x}$
* $y = x^3$
* $y = 2^x$
* $y = 1 / (x + 1)$


You can use functions from the math module to help implement some of the equations above.



```python
#create an array with positive numbers only
x1 = np.arange(0, 50)

# in my graphs for equations 1 and 4 i'm using x1, and for equations 2 and 3 - x
```

```python
y1 = [np.sqrt(i) for i in x1]
y2 = [i ** 3 for i in x]
y3 = [2 ** i for i in x]
y4 = [(1 /(i + 1)) for i in x1]
```

```python
plt.plot(x1, y1,color = 'r')
plt.title('$y = \sqrt{x}$')
plt.grid(True, ls = '--')
plt.show()
```

```python
plt.plot(x, y2,color = 'b', alpha = 0.8)
plt.title('$y = x^3$')
plt.grid(True, ls = '--')
plt.xticks(range(-10, 11, 2))
plt.show()
```

```python
plt.plot(x, y3,color = 'g', alpha = 0.9)
plt.title('$y = 2^x$')
plt.grid(True, ls = '--')
plt.xticks(range(-10, 11, 2))
plt.show()
```

```python
plt.plot(x1, y4,color = 'm', ls = '--', alpha = 0.5)
plt.title('$y = 1 / (x+ 1)$')
plt.show()
```

**3. Combine the figures you created in the last step into one large figure with 4 subplots.**

```python
#use x1 and y1, y2, y3, y4 from the previous exercise

plt.figure(figsize = (16, 8))

#first subplot for x1 and y1
plt.subplot(221)
plt.plot(x1, y1,color = 'r', ls = ':')
plt.title('$y = \sqrt{x}$')

plt.subplot(222)
plt.plot(x, y2,color = 'b', ls = ':', alpha = 0.8)
plt.title('$y = x^3$')

plt.subplot(223)
plt.plot(x, y3,color = 'g', ls = ':', alpha = 0.9)
plt.title('$y = 2^x$')

plt.subplot(224)
plt.plot(x1, y4,color = 'm', ls = '--', alpha = 0.5)
plt.title('$y = 1\frac{(x+ 1)}$')

plt.suptitle('Equations')
plt.show()
```
